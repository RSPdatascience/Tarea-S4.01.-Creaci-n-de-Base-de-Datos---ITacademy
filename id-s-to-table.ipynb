```python
import pandas as pd

transactions = pd.read_csv('transactions.csv', delimiter=';')
transactions
```

```python
# Dividir en filas los valores separados por coma de la columna product_ids 
transactions_split = transactions['product_ids'].str.split(',', expand=True)
transactions_split
```

```python
# Insertar los transaction_id en la tabla
transactions_split.insert(0, 'transaction_id', transactions['id'])
transactions_split
```

```python
# Coonvertir las columnas en filas, manteniendo la variable 'transaction_id' como identificador de cada fila y 'product_id' como el valor correspondiente al 'transaction_id'
products_transactions = transactions_split.melt(id_vars='transaction_id', value_name='product_id').dropna()
products_transactions
```

```python
# Elimanr la columna 'variable' generada por la funciÃ³n melt
products_transactions.drop(columns=['variable'], inplace=True)
```

```python
products_transactions.to_csv('products_transactions.csv', index=False)
```
